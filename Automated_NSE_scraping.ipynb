```python
import datetime
from nsedt import equity as eq
from datetime import date, timedelta
import os
import pandas as pd

# Get today's date


# Rest of the code remains the same
symbols= ['CIPLA','INDUSINDBK','TECHM','WIPRO','BHARTIARTL','AXISBANK',
          'HCLTECH','HDFCBANK','HDFCLIFE','RELIANCE','LTIM','TCS','LT','INFY','SBILIFE',
          'HINDALCO','KOTAKBANK','TATASTEEL','UPL','ONGC']

# Create an empty dictionary to store the dataframes

drive_folder = '/content/drive/My Drive/stock_data'

for symbol in symbols:
  file_path = os.path.join(drive_folder, f"{symbol}.csv")
  df=pd.read_csv(file_path)
  data = eq.get_price(today_date, tomorrow_date, symbol=symbol)
  df = df.append(data, ignore_index=True)
  df.to_csv(file_path, index=False)
  print(symbol,"Done")
```
