```python
import os

def get_full_paths_from_folder(folder_path):
    try:
        # Check if the given path is a valid directory
        if not os.path.isdir(folder_path):
            raise NotADirectoryError(f"{folder_path} is not a valid directory.")

        # Get a list of all filenames in the folder
        filenames = os.listdir(folder_path)

        # Get full paths by joining the folder path with each filename
        full_paths = [os.path.join(folder_path, filename) for filename in filenames]

        return full_paths
    except Exception as e:
        print(f"Error: {e}")
        return []

# Replace 'folder_path' with the path of the folder you want to get filenames from
folder_path = '/content/drive/MyDrive/stock_data'
stock_files = get_full_paths_from_folder(folder_path)
stock_files

```

```python
import pandas as pd

def calculate_rsi(data, window=14):
    delta = data.diff()
    gain = (delta.where(delta > 0, 0)).fillna(0)
    loss = (-delta.where(delta < 0, 0)).fillna(0)

    avg_gain = gain.rolling(window=window).mean()
    avg_loss = loss.rolling(window=window).mean()

    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))

    return rsi

def calculate_bollinger_bands(data, window=20, num_std=2):
    rolling_mean = data.rolling(window=window).mean()
    rolling_std = data.rolling(window=window).std()

    upper_band = rolling_mean + (rolling_std * num_std)
    lower_band = rolling_mean - (rolling_std * num_std)

    return upper_band, rolling_mean, lower_band

def calculate_stochastic_oscillator(high, low, close, window=14):
    lowest_low = low.rolling(window=window).min()
    highest_high = high.rolling(window=window).max()

    K = ((close - lowest_low) / (highest_high - lowest_low)) * 100
    D = K.rolling(window=3).mean()

    return K, D

def calculate_macd(data, fast_period=12, slow_period=26, signal_period=9):
    ema_fast = data.ewm(span=fast_period).mean()
    ema_slow = data.ewm(span=slow_period).mean()

    macd = ema_fast - ema_slow
    signal_line = macd.ewm(span=signal_period).mean()

    return macd, signal_line

def calculate_ma(data, window=20):
    ma = data.rolling(window=window).mean()
    return ma

for file in stock_files:
  df=pd.read_csv(file)
  df['RSI'] = calculate_rsi(df['Close Price'])
  df['Upper Band'], df['Middle Band'], df['Lower Band'] = calculate_bollinger_bands(df['Close Price'])
  df['%K'], df['%D'] = calculate_stochastic_oscillator(df['High Price'], df['Low Price'], df['Close Price'])
  df['MACD'], df['Signal Line'] = calculate_macd(df['Close Price'])
  df['MA'] = calculate_ma(df['Close Price'])
  df.to_csv(file)


  # Display the modified DataFrame
```

```python
# Assuming df is your DataFrame with all the indicator columns
for file in stock_files:
  df=pd.read_csv(file)
# Define thresholds for each indicator
  rsi_overbought_threshold = 70
  rsi_oversold_threshold = 30
  bollinger_sell_threshold = df['Close Price'].mean() + 2 * df['Close Price'].std()
  bollinger_buy_threshold = df['Close Price'].mean() - 2 * df['Close Price'].std()
  stochastic_sell_threshold = 80
  stochastic_buy_threshold = 20
  macd_sell_threshold = 0
  macd_buy_threshold = 0
  ma_sell_threshold = df['Close Price'].mean() - 0.02 * df['Close Price'].mean()  # 2% below the moving average
  ma_buy_threshold = df['Close Price'].mean() + 0.02 * df['Close Price'].mean()  # 2% above the moving average

  # Calculate recommendations
  df['RSI Recommendation'] = 'Hold'
  df.loc[df['RSI'] > rsi_overbought_threshold, 'RSI Recommendation'] = 'Sell'
  df.loc[df['RSI'] < rsi_oversold_threshold, 'RSI Recommendation'] = 'Buy'

  df['Bollinger Recommendation'] = 'Hold'
  df.loc[df['Close Price'] > bollinger_sell_threshold, 'Bollinger Recommendation'] = 'Sell'
  df.loc[df['Close Price'] < bollinger_buy_threshold, 'Bollinger Recommendation'] = 'Buy'

  df['Stochastic Recommendation'] = 'Hold'
  df.loc[df['%K'] > stochastic_sell_threshold, 'Stochastic Recommendation'] = 'Sell'
  df.loc[df['%K'] < stochastic_buy_threshold, 'Stochastic Recommendation'] = 'Buy'

  df['MACD Recommendation'] = 'Hold'
  df.loc[df['MACD'] > macd_sell_threshold, 'MACD Recommendation'] = 'Sell'
  df.loc[df['MACD'] < macd_buy_threshold, 'MACD Recommendation'] = 'Buy'

  df['MA Recommendation'] = 'Hold'
  df.loc[df['Close Price'] > ma_sell_threshold, 'MA Recommendation'] = 'Sell'
  df.loc[df['Close Price'] < ma_buy_threshold, 'MA Recommendation'] = 'Buy'

  df.to_csv(file)


```
