```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from keras.models import Model
from keras.layers import Input, LSTM, Dense, Attention, Concatenate
from keras.optimizers import Adam
from keras.callbacks import LearningRateScheduler, EarlyStopping, ModelCheckpoint

# Load your historical stock price data into a pandas DataFrame
# Assuming 'data' is your DataFrame containing the historical data

# Function to create sequences for training
def create_sequences(data, seq_length):
    sequences = []
    for i in range(len(data) - seq_length):
        sequences.append(data[i:i+seq_length])
    return np.array(sequences)


# Define the encoder-decoder model
def build_model(input_shape, output_shape):
    inputs = Input(shape=input_shape)
    encoder = LSTM(units=128, return_sequences=True)(inputs)
    encoder = LSTM(units=128, return_sequences=True)(encoder)
    encoder = LSTM(units=128, return_sequences=True)(encoder)

    decoder_input = Input(shape=output_shape)
    decoder = LSTM(units=128, return_sequences=True)(decoder_input)
    decoder = LSTM(units=128, return_sequences=True)(decoder)

    attention = Attention()([decoder, encoder])
    merged = Concatenate()([decoder, attention])
    merged = Dense(128, activation='relu')(merged)

    output = Dense(output_shape[1])(merged)

    model = Model(inputs=[inputs, decoder_input], outputs=output)
    return model



# Define a learning rate schedule
def lr_schedule(epoch):
    if epoch < 50:
        return 0.001
    elif epoch < 100:
        return 0.0005
    else:
        return 0.0001

# Custom metric for accuracy
def custom_accuracy(y_true, y_pred):
    threshold = 0.1  # Adjust as needed
    absolute_diff = np.abs(y_true - y_pred)
    correct_predictions = np.where(absolute_diff <= threshold, 1, 0)
    return np.mean(correct_predictions)


symbols= ['CIPLA','INDUSINDBK','TECHM','WIPRO','BHARTIARTL','AXISBANK',
          'HCLTECH','HDFCBANK','HDFCLIFE','RELIANCE','LTIM','TCS','LT','INFY','SBILIFE',
          'HINDALCO','KOTAKBANK','TATASTEEL','UPL','ONGC']
drive_folder = '/content/drive/My Drive/stock_data'

i=0
# Iterate through stock files
for symbol in symbols:
    file_path = os.path.join(drive_folder, f"{symbol}.csv")
    data = pd.read_csv(file_path)

    # Select relevant columns
    data.drop(columns=["Date",'Unnamed: 0.1', 'Unnamed: 0.2','Unnamed: 0.3','Unnamed: 0', "Prev Close Price", "52 Week High Price", "52 Week Low Price",
                       'Deliverable Volume', 'Deliverable Percent', 'Series','RSI',
       'Upper Band', 'Middle Band', 'Lower Band', '%K', '%D', 'MACD',
       'Signal Line', 'MA', 'RSI Recommendation', 'Bollinger Recommendation',
       'Stochastic Recommendation', 'MACD Recommendation',
       'MA Recommendation'], inplace=True)
    print(data.shape)
    print(data.columns)

    # Normalize the data
    scaler = StandardScaler()
    data_scaled = scaler.fit_transform(data)

    # Set the sequence length
    sequence_length = 60  # You can change this value

    # Create sequences for training
    sequences = create_sequences(data_scaled, sequence_length)

    # Split the data into features and target
    X = sequences[:, :-1]
    y = sequences[:, -1]

    # Split data into training and testing sets
    split_ratio = 0.8  # Adjust as needed
    split_index = int(split_ratio * len(X))
    X_train, X_test = X[:split_index], X[split_index:]
    y_train, y_test = y[:split_index], y[split_index:]
    y_test=y_test.reshape(y_test.shape[0],1,7)
    y_train=y_train.reshape(y_train.shape[0],1,7)

    # Build the encoder-decoder model
    model = build_model(input_shape=(X_train.shape[1], X_train.shape[2]), output_shape=(1,y_train.shape[2]))

    # Compile the model
    optimizer = Adam(learning_rate=0.001)
    model.compile(optimizer=optimizer, loss='mse')

    # Define callbacks
    lr_scheduler = LearningRateScheduler(lr_schedule)
    early_stopping = EarlyStopping(patience=10, restore_best_weights=True)
    model_checkpoint = ModelCheckpoint(f'trial', save_best_only=True)

    # Train the model with callbacks
    history = model.fit([X_train, y_train], y_train, epochs=100, batch_size=64, validation_data=([X_test, y_test], y_test),
                        callbacks=[lr_scheduler, early_stopping,model_checkpoint])
    model.save("/content/drive/MyDrive/stock_models/model_23 11/{}.h5".format(symbols[i]))


    # Print validation loss and accuracy at each epoch
    for epoch in range(len(history.history['val_loss'])):
        val_loss = history.history['val_loss'][epoch]
        val_acc = custom_accuracy(y_test, model.predict([X_test, y_test]))
        print(f"File: {file}, Epoch {epoch+1}: Val Loss = {val_loss:.4f}, Val Accuracy = {val_acc:.4f}")
    print(symbols[i]," is saved")
    i=i+1
```
